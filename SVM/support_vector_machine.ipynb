```python
# packages for data analysis
import numpy as np
import pandas as pd 

from sklearn import svm

import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)

%matplotlib inline
```

```python
recipes = pd.read_csv("Cupcakes vs Muffins.csv")
print(recipes.head())
```

```python
# plot our data
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s":70})
```

```python
# format or pre process our data
type_label = np.where(recipes['Type']=="Muffin", 0, 1)
recipe_features = recipes.columns.values[1:].tolist()
recipe_features
ingredients = recipes[["Flour", "Sugar"]].values
print(ingredients)
```

```python
# fit model
model = svm.SVC(kernel='linear')
model.fit(ingredients, type_label)
```

```python
# get the separating hyperplane
w = model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(30, 60)
yy = a * xx - (model.intercept_[0]) / w[1]

# plot the parallels to the seperating hyperplane that pass through the support vectors
b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```

```python
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(xx,yy_down, 'k--')
plt.plot(xx,yy_up, 'k--')
```

```python
# create a functino to predict muffin or cupcake
def muffin_or_cupcake(flour, sugar):
  if(model.predict([[flour, sugar]]))==0:
    print("You are looking at a muffin recipe!")
  else:
    print("You are looking at a cupcake recipe!")

# predict if 50 parts flour and 20 parts sugar)
muffin_or_cupcake(50, 20)  
```

```python
# Let's plot this one the graph
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s":70})
plt.plot(xx,yy, linewidth=2, color='black')
plt.plot(50, 20, 'yo', markersize='9')
```
